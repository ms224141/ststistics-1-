```python
#    Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis? Task: Write Python code to calculate the z-score and p-value for the given data. Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.


Sample mean  = 105
Population mean  = 100
Population standard deviation  = 15
Sample size  = 25
Significance level = 0.05
Step 1: Calculate the Z-score
The formula for the z-score is:

z = (Sample mean  - Population mean) / (standard deviation / √n)
Substituting the values:

z = (105 - 100) / (15 / √25)
z = 5 / 3
z ≈ 1.67
Step 2: Calculate the P-value
We can use a z-table or a statistical software to find the p-value corresponding to a z-score of 1.67.

Since we're conducting a two-tailed test (we're interested in deviations from the mean in both directions), we need to find the area in both tails.

Using a z-table or statistical software, we find that the area to the right of z = 1.67 is approximately 0.0475.

Therefore, the p-value for a two-tailed test is:

p-value = 2 * 0.0475 = 0.095
Step 3: Decision
We compare the p-value to the significance level (alpha = 0.05):

If p-value < alpha, we reject the null hypothesis.
If p-value ≥ alpha, we fail to reject the null hypothesis.
In this case, 0.095 ≥ 0.05, so we fail to reject the null hypothesis.

Conclusion:
Based on the given information and a significance level of 0.05, there is not enough evidence to conclude that the sample mean is significantly different from the population mean.





```

```python
#    Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the distribution. What is the expected mean and variance? Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance. Objective: Understand the properties of a binomial distribution and verify them through simulation.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of trials
p = 0.6  # Probability of success
num_samples = 1000

# Generate 1000 samples from the binomial distribution
samples = np.random.binomial(n, p, num_samples)

# Calculate the expected mean and variance
expected_mean = n * p
expected_variance = n * p * (1-p)

# Plot the histogram of the samples
plt.hist(samples, bins=range(n+2), align='left', rwidth=0.8)
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.show()

# Print the expected and sample mean and variance
print("Expected mean:", expected_mean)
print("Sample mean:", np.mean(samples))
print("Expected variance:", expected_variance)
print("Sample variance:", np.var(samples))
```
